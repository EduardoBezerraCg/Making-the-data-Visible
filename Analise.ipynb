```python
#Come√ßamos por importar nossas bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importando nossa BD
dev_mun = pd.read_csv('dados.csv', sep=';')
dev_mun.head()
```

```python
#Setar o tema
sns.set_theme(style='darkgrid')
```

```python
cinco_maiores = dev_mun.sort_values(by='PIB', ascending=False).head(5)
cinco_maiores
```

```python
#Criando o gr√°fico mostrando os 5 maiores
sns.barplot(x='PIB', y='MUNICIPIO', data=cinco_maiores)
plt.title('5 maiores PIBS')
```

```python
#Obtendo os 5 menores
cinco_menores = dev_mun.sort_values(by=['PIB']).head(6)
cinco_menores
```

```python
sns.barplot(x='PIB', y='MUNICIPIO', data=cinco_menores)
plt.title('Os 5 menores')
plt.tight_layout()
```

```python
sns.histplot(dev_mun.VALOREMPENHO, kde=True, bins=10).set_title('VALOREMPENHO')
```

```python
#Calculando os 10 maiores
agrupado = dev_mun.sort_values(['VALOREMPENHO']).head(10)
agrupado = agrupado.iloc[:,[1,3]]
agrupado.plot.bar(x='MUNICIPIO', y='VALOREMPENHO', color='LightBlue')
plt.title('Top 10 Munic√≠pios com maior valor empenho')
```

```python
#
data1 = dev_mun.groupby(['VALOREMPENHO'])['MUNICIPIO'].sum()
data1
```

```python
sns.histplot(data=dev_mun['PIB'], color='LightBlue', kde=True, stat='density', bins=15).set_title('PIB')
```

```python
#entendendo o PIB atrav√©s do gr√°fico de dispers√£o
plt.scatter(x=dev_mun.MUNICIPIO.head(20), y=dev_mun.PIB.head(20), color='Brown', facecolors = 'none', marker='.')
plt.xlabel('Cidades')
plt.xticks(rotation=80)
plt.tight_layout
```

```python
#Experimentando integrar os 5 maiores apenas neste grafico
agrupado = dev_mun.sort_values(by=['PIB'], ascending=False).head(20)
agrupado
```

```python
top5 = dev_mun['PIB'].nlargest(5)
```

```python
plt.scatter(x='PIB', y='MUNICIPIO', data=agrupado)
```

```python

plt.plot(top5, marker='.')
```

```python
#Nesse caso √© mais dificil criar uma linha de tend√™ncia devido a dispers√£o dos dados
sns.regplot(x=dev_mun.PIB.head(10), y=dev_mun.VALOREMPENHO.head(10), data=dev_mun, x_jitter=0.3, fit_reg=True)
```

```python
#Compilando o que analizamos
plt.figure(1)
plt.subplot(2,2,1)
plt.ylabel('Plot Graphic')
plt.scatter(dev_mun.PIB, dev_mun.VALOREMPENHO)

plt.subplot(2,2,2)
plt.scatter(x='PIB', y='MUNICIPIO', data=agrupado.head(10))
plt.ylabel('Scatter Graphic')

plt.subplot(2,2,3)
plt.plot(top5, marker='.')
plt.ylabel('Line Graphic')

plt.subplot(2,2,4)
sns.histplot(dev_mun.VALOREMPENHO, kde=True, bins=10)
plt.ylabel('Valor Empenho')

plt.tight_layout()
```
